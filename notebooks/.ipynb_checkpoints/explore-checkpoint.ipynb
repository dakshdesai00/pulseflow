```python
# ============================================================
# FINAL EXPLORATION SUMMARY
# ============================================================

print("=== DATASET OVERVIEW ===")
print(f"Total transactions: {len(df):,}")
print(f"Fraud transactions: {df['isFraud'].sum():,} ({df['isFraud'].mean()*100:.2f}%)")
print(f"Unique merchants (card1): {df['card1'].nunique():,}")

print("\n=== FRAUD RATE BY PRODUCT ===")
print(df.groupby('ProductCD')['isFraud'].agg(['sum','count','mean'])
      .rename(columns={'sum':'fraud_count','count':'total','mean':'fraud_rate'})
      .sort_values('fraud_rate', ascending=False))

print("\n=== AVG TRANSACTION AMOUNT ===")
print(f"Non-fraud: ${df[df['isFraud']==0]['TransactionAmt'].mean():.2f}")
print(f"Fraud:     ${df[df['isFraud']==1]['TransactionAmt'].mean():.2f}")

print("\n=== ANONYMOUS EMAIL TRANSACTIONS ===")
anon = df[df['P_emaildomain'] == 'anonymous.com']
print(f"Total anonymous email txns: {len(anon):,}")
print(f"Fraud rate in anonymous:    {anon['isFraud'].mean()*100:.2f}%")
print(f"Fraud rate overall:         {df['isFraud'].mean()*100:.2f}%")

print("\n=== CARD TYPE DISTRIBUTION ===")
print(df.groupby('card6')['isFraud'].agg(['sum','count','mean'])
      .rename(columns={'sum':'fraud_count','count':'total','mean':'fraud_rate'}))

print("\n=== MERCHANT TRANSACTION DISTRIBUTION ===")
merchant_counts = df.groupby('card1')['TransactionID'].count()
print(merchant_counts.describe())
print(f"\nMerchants with < 5 txns (will filter): {(merchant_counts < 5).sum():,}")
print(f"Merchants with >= 5 txns (will keep):  {(merchant_counts >= 5).sum():,}")

print("\n=== KEY INSIGHTS ===")
print("1. Product C has highest fraud rate (~11.7%)")
print("2. Fraud transactions avg $15 higher than non-fraud")
print("3. anonymous.com email is a strong fraud signal")
print("4. 75% of merchants have <= 14 transactions — filter < 5 to reduce noise")
print("5. Low transaction count merchants show 100% fraud rate — statistically meaningless")
print("6. TransactionDT mean higher in fraud — fraud concentrated later in time window")
```
